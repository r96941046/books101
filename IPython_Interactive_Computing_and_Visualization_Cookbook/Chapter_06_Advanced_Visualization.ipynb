#### prettyplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
np.random.seed(12)
```

```python
for i in range(8):
    x = np.arange(1000)
    y = np.random.randn(1000).cumsum()
    plt.plot(x, y, label=str(i))
    plt.legend()
```

```python
import prettyplotlib as ppl
```

```python
np.random.seed(12)
for i in range(8):
    x = np.arange(1000)
    y = np.random.randn(1000).cumsum()
    ppl.plot(x, y, label=str(i))
    ppl.legend()
```

```python
x1 = np.random.randn(80)
x2 = np.random.randn(80)
x3 = x1 * x2
```

```python
y1 = 0.5 + 2 * x1 - x2 + 2.5 * x3 + 3 * np.random.randn(80)
y2 = 0.5 + 2 * x1 - x2 + 2.5 * np.random.randn(80)
y3 = y2 + np.random.randn(80)
```

```python
sns.violinplot([x1, x2, x3])
```

```python
sns.regplot(x2, y2)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dict(x1=x1, x2=x2, x3=x3, y1=y1, y2=y2, y3=y3))
```

```python
sns.corrplot(df)
```

```python
import numpy as np
import bokeh.plotting as bkh
from bokeh.io import output_notebook
output_notebook()
```

```python
x = np.linspace(0.0, 1.0, 100)
y = np.cumsum(np.random.randn(100))
```

```python
f = bkh.figure()
f.line(x, y, line_width=2)
bkh.show(f)
```
